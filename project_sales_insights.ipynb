```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('sales_data.csv', encoding="ISO-8859-1")
df
```

```python
df.info()
```

#fetching the null values from the amount column

```python
null_values=df[df["Amount"].isnull()]
print(null_values)
```

#Droping unwanted columns such as staus, unnamed1


```python
df.drop(["Status","unnamed1"], axis=1, inplace=True)
df.info()
```

#Delete all nan values from the data frame


```python
cleaned_data=df.dropna()
print(cleaned_data)
```

#Rename the column Marital_Status to Marital


```python
d={'Marital_Status':"Marital"}
df.rename(columns=d, inplace=True)
df
```

change the values of Marital col 0-->single, 1--> married
**italicized text**

```python
m={0:"single",1:"married"}
df["Marital"]=df['Marital'].map(m)
df
# df.drop("'Marital", axis=1, inplace=True)
# df
```

#change the values of Gender col F-->Female, M-->Male


```python
sex={"F":"Female","M":"Male"}
df['Gender']=df['Gender'].map(sex)
#df['Gender'].isnull()
df
```

visualize the count of males and females based on gender column using
seaborn

```python
ax=sns.countplot(x='Gender',data=df,hue="Marital")
for i in ax.containers:
  ax.bar_label(i)
```

**conclusion** count of females are more in both the cases

```python
sales=df.groupby(['Gender'])['Amount'].sum()
```

group by gender, age_group,
state,Occupation,Product_category,marital_status

```python
sales=df.groupby(['Gender'])["Amount"].sum().reset_index()
ax=sns.barplot(x="Gender",y="Amount", data=sales,hue="Gender")
for i in ax.containers:
  ax.bar_label(i)
```

age_group

```python
ax=sns.countplot(x="Age Group",data= df, hue="Marital")
for i in ax.containers:
  ax.bar_label(i)
```

```python
#total amount vs age_group
sales_age=df.groupby(df["Age Group"])['Amount'].sum().reset_index()
sns.barplot(x="Age Group", y ="Amount", data= sales_age,  palette="dark:pink")
```

from the above graph we can see that more no of amount has been spend by the group bet (26-35)

**state**


```python
#total amount of money spend by statewise
a=df.groupby('State')['Amount'].sum().reset_index()
plt.figure(figsize=(20,10))
state=sns.barplot(x="State", y="Amount",data= a, color="green",palette="dark:green")
#sns.set(rc={figure.figsize(20,10)})
for i in state.containers:
  state.bar_label(i)
```

conclusion:- from the above graph we can understand that up,  has spent more amount of money, then maharastra, karnataka and so on


```python
#rotal amount of money apend by occupation Occupation
a=df.groupby(df['Occupation'])['Amount'].sum().reset_index()
plt.figure(figsize=(20,10))
ax=sns.barplot(x="Occupation",y="Amount", data= a, hue="Occupation" )
for i in ax.containers:
  ax.bar_label(i)
```

conclusion:- those who are working in IT sector is spending more amount of money, then healthcare, then aviation and so on

Product_category

```python
plt.figure(figsize=(25,10))
ax=sns.countplot(x='Product_Category', data = df, hue="Gender")
for i in ax.containers:
  ax.bar_label(i)
```

```python
b=df.groupby('Product_Category',as_index=False)['Amount'].sum()
b=pd.DataFrame(b)
plt.figure(figsize=(25,10))
ax=sns.barplot(x="Product_Category", y="Amount", data=b, hue="Product_Category")
for i in ax.containers:
  ax.bar_label(i)
```

conclusion:-more no of females are spending more on food,clothing & apparel, footware and shoes and electronic gadgets.

```python
#marital_status
ax=sns.countplot(x="Marital", data= df, hue="Gender")
for i in ax.containers:
  ax.bar_label(i)
```

```python
m=df.groupby(['Marital','Gender'])['Amount'].sum()
M=pd.DataFrame(m)
plt.figure(figsize=(10,10))
ax=sns.barplot(x="Marital",y="Amount", data= M, hue="Gender")
for i in ax.containers:
  ax.bar_label(i, fmt= "%0.1f")
```

conclusion:- females those who are single are spending more amount

the overall **conclusion** is females are more and spending more money those who are single betwwen the age_group of 26-35, and working in IT, healthcare and aviation, and mostly spending on Food, clothing & apparel, footware and shoes and electronic gadgets, and from states UP, maharastra, karnataka.
